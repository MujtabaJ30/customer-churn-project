#### Data Loading, Merging, and Cleaning

```python
import pandas as pd
```

```python
df_demographics = pd.read_excel('..\\data\\raw\\Telco_customer_churn_demographics.xlsx')
df_location = pd.read_excel('..\\data\\raw\\Telco_customer_churn_location.xlsx')
df_population = pd.read_excel('..\\data\\raw\\Telco_customer_churn_population.xlsx')
df_services = pd.read_excel('..\\data\\raw\\Telco_customer_churn_services.xlsx')
df_status = pd.read_excel('..\\data\\raw\\Telco_customer_churn_status.xlsx')

print("Demographics:", df_demographics.shape)
print("Location:", df_location.shape)
print("Population:", df_population.shape)
print("Services:", df_services.shape)
print("Status:", df_status.shape)
```

```python
df_demographics = df_demographics.drop(columns=['Count'])
df_services = df_services.drop(columns=['Count', 'Quarter'])
df_status = df_status.drop(columns=['Count', 'Quarter'])
df_location = df_location.drop(columns=['Count'])

df = df_demographics.merge(df_services, on='Customer ID', how='left')
df = df.merge(df_status, on='Customer ID', how='left')
df = df.merge(df_location, on='Customer ID', how='left')

print(df.shape)
```

```python
df.head()
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df['Offer'] = df['Offer'].fillna('No Offer')
df['Internet Type'] = df['Internet Type'].fillna('No Internet')
df['Churn Category'] = df['Churn Category'].fillna('No Churn')
df['Churn Reason'] = df['Churn Reason'].fillna('No Churn')

df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df.to_csv('..\\data\\processed\\master_df.csv', index=False)
```
